```python
import pandas as pd
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
```

```python
table = data[0]
table.columns = table.iloc[0]
#table=table[["Security","Symbol","GICS Sector"]]
table=table[1:]
```

```python
path='/Users/hemanth/Desktop/test2/socgen/'
mark=pd.read_excel(path+'sp.xlsx')
mark=mark.rename(columns={"Ticker" :"Symbol"})
```

```python
ticker=pd.merge(table,mark,on=["Symbol"],how="left")
```

```python
ticker=ticker[['Security',"Symbol","GICS Sector","GICS Sub Industry","Dividend Yield","Market Cap ($M)","P/E Ratio"]]
```

```python
ticker.to_csv(path+"tickers.csv",index=False)
```
